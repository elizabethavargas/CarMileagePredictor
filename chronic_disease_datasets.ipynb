# Create Chronic Disease Datasets
data from https://catalog.data.gov/dataset/u-s-chronic-disease-indicators-cdi

```python
# imports
import pandas as pd
```

```python
# read file
chronic_disease = pd.read_csv('U.S._Chronic_Disease_Indicators__CDI_.csv')
chronic_disease
```

```python
# filter for asthma
asthma = chronic_disease[chronic_disease.Topic == 'Asthma']

# filter for alcohol
alcohol = chronic_disease[chronic_disease.Topic == 'Alcohol']

# filter for alcohol
tobacco = chronic_disease[chronic_disease.Topic == 'Tobacco']
```

```python
# save files
asthma.to_csv('asthma_stats.csv')
alcohol.to_csv('alcohol_stats.csv')
tobacco.to_csv('tobacco_stats.csv')

# download files
from google.colab import files
files.download('asthma_stats.csv') 
files.download('alcohol_stats.csv') 
files.download('tobacco_stats.csv') 
```
